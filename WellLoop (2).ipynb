##🔥 4. Embedding Modelini Yükleme

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

## ❓ 4.1 Soru Metinlerini Embedding İşlemi

```python
question_texts = [f"Question: {row['questionText']}" for _, row in df.iterrows()]
```

```python
embeddings = model.encode(question_texts, convert_to_tensor=False)
```

## 🔢 5. FAISS Indexleme ve Embedding Ekleme İşlemi

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings).astype('float32'))
```

## 🔍 7. Sorguya En Benzer Cevapları Bulma

```python
def retrieve_top_k_responses(query, k=5):
    question_texts = [f"Question: {row['questionText']}" for _, row in df.iterrows()]
    answer_texts = [f"Answer: {row['answerText']}" for _, row in df.iterrows()]

    query_embedding = model.encode(query, convert_to_tensor=False)
    query_embedding = np.array(query_embedding).reshape(1, -1).astype('float32')

    D, I = index.search(query_embedding, k)

    top_answers = [answer_texts[i] for i in I[0]]

    return top_answers
```

## 🔄🤖 8. Kullanıcı Sorgusu ve Gemini API ile Yanıt Oluşturma

```python
def query_gemini_with_prompt(api_key, user_query, retrieved_responses, prompt, lang):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key={api_key}"

    headers = {"Content-Type": "application/json"}

    history_context = "\n".join(chat_history[-5:])
    context = "\n\n".join(retrieved_responses)

    language_instruction = f"Please answer in {lang}."

    full_prompt = f"""
{prompt}

{language_instruction}

### Chat History:
{history_context}

### Retrieved Answers:
{context}

### User Query:
{user_query}
    """

    data = {"contents": [{"parts": [{"text": full_prompt}]}]}

    response = requests.post(url, headers=headers, json=data)

    #print("\n🔍 Gemini API Response JSON:", response.json())

    if response.status_code == 200:
        try:
            result = response.json()
            if 'candidates' in result and len(result['candidates']) > 0:
                response_text = result['candidates'][0]['content']['parts'][0]['text']
                return response_text
            else:
                ##print("🚨 Expected 'candidates' key is missing or empty in the response.")
                return None
        except Exception as e:
            #print(f"🚨 Error processing the response: {e}")
            return None

    else:
        #print(f"🚨 API Error {response.status_code}: {response.text}")
        return None

custom_prompt = """
You are an AI assistant specializing in mental health counseling.
Behave like a psychologist. Try to UNDERSTAND EMOTIONS!. Ask questions to help users explore their feelings and thoughts.
Detect situations where the user is at risk of harming themselves and recommend emergency help lines when necessary.
Recognize even small achievements of the user and use motivating language.
Use a friendly, conversational tone, like a close friend giving advice.
Use motivating language.
Focus more on listening to the user.
Don't jump from topic to topic!
If the user is experiencing a severe psychological problem, refer them to a specialist.
If the user is extremely stressed or inconsolable, help defuse the situation by responding quickly and calmly.
DON'T give long answers.
"""

chat_history = []
translator = Translator()
flag = True

while True:
    if (flag):
      user_query = "Psikolojik destek asistanı gibi selamla. Sıcak, samimi ve kısa bir selam ver. Konuşmaya başlaması için motive edici bir dil kullan."
      flag = False
    else:
      user_query = input("\n👤 You: ")

    detected_lang = translator.detect(user_query).lang

    translated = translator.translate(user_query, src=detected_lang, dest='en')
    user_query = translated.text

    if user_query.lower() in ["exit", "quit", "bye"]:
        print("🤖 Chatbot: Goodbye! Take care. 😊")#buraya da flagle konusmayı bıtırdıgı ve bu adımı yaptıgı için tebrik eden bir cevap
        break

    retrieved_responses = retrieve_top_k_responses(user_query, k=5)

    bot_response = query_gemini_with_prompt(os.environ["GEMINI_API_KEY"], user_query, retrieved_responses, custom_prompt, detected_lang)

    chat_history.append(f"You: {user_query}")
    chat_history.append(f"Chatbot: {bot_response}")

    print(f"\n🤖 Chatbot: {bot_response}")
```
